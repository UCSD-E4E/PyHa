```python
import os
import pandas as pd
audio_dir = "./gt/"
```

```python
df_cols = ['FOLDER', 'IN FILE' , 'CLIP LENGTH', 'CHANNEL', 'OFFSET', 'DURATION', 'SAMPLE RATE', 'MANUAL ID']
manual_df = pd.DataFrame(columns=df_cols)
```

```python
def split_time_x(time):
    time = time.split('-')[0]
    time = time.split(':')
    #print(time)
    time = int(time[2])
    return time

def split_time_y(time):
    time = time.split('-')[0]
    time = time.split(':')
    #print(time)
    time = int(time[1])
    return time


    
manual_df = pd.DataFrame(columns=df_cols)

for file in os.listdir(audio_dir):
    try:
        filename, extension = file.split('.')
    except:
        pass
    
    if (extension == "ipynb_checkpoints"):
        continue
        
    def add_file_name(num):
        return filename + "_" + str(num) + "."+ "wav"

    print(file)
    file_df = pd.read_csv(audio_dir+file, header=None)
    file_df.rename(columns = {0:'time', 1:'MANUAL ID'}, inplace = True)
    file_df["OFFSET"] = file_df["time"].apply(split_time_x)
    file_df["CLIP_OFFSET"] = file_df["time"].apply(split_time_y)
    file_df["DURATION"] = 5.0
    file_df["FOLDER"] = "./BirdCLEF2020_Validation_Audio_and_Ground_Truth/audio/audio_min/"
    file_df["IN FILE"] =  file_df["CLIP_OFFSET"].apply(add_file_name) 
    file_df["CLIP LENGTH"] = 60
    file_df["CHANNEL"] = 1
    file_df["SAMPLE RATE"] = 48000
    file_df = file_df[["FOLDER", "IN FILE", "CLIP LENGTH","CHANNEL", "SAMPLE RATE", "OFFSET","DURATION", 'MANUAL ID' ]]
    manual_df = manual_df.append(file_df)
    manual_df = manual_df.reset_index(drop=True)
manual_df
```

```python
manual_df.to_csv("BIRDCLEF2020_SPECIES_ID.csv")
```
