```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

import pandas as pd
import numpy as np
from PyHa.statistics import *
from PyHa.IsoAutio import *
from PyHa.visualizations import *
```

```python
#path_to_audio_files = "/home/jacob/acoustic-species-id/Mixed_Bird/"
#path_to_ground_truth = "MDD_Xeno_Canto_DSC180_Labels_uniform_3s_binary.csv"
chunk_length = 3
path_to_audio_files = "./mixed_bird/025_Mixed_Bird_Outputs_Train/"#"./TEST/"#"./mixed_bird/025_Mixed_Bird_Outputs_Train/"#"./mixed_bird/Mixed_Bird/"
path_to_ground_truth = "./mixed_bird/mixed_bird_manual.csv" #"ScreamingPiha_Manual_Labels.csv"#"./mixed_bird/mixed_bird_manual.csv"
```

```python
manual_df = pd.read_csv(path_to_ground_truth)
manual_df["MANUAL ID"] = "bird"
manual_df_chunked = annotation_chunker(manual_df, chunk_length)
```

```python
# Parameters to define isolation behavior
isolation_parameters_micro = {
    "model" : "microfaune",
    "technique" : "chunk",
    "threshold_type" : "median",
    "threshold_const" : 4.0,
    "threshold_min" : 0.25,
    "window_size" : 2.0,
    "chunk_size" : 3.0
}
automated_df_micro = generate_automated_labels(path_to_audio_files,isolation_parameters_micro);
automated_df_micro["MANUAL ID"] = "bird"
automated_df_micro_chunked = annotation_chunker(automated_df_micro, chunk_length)
```

```python
isolation_parameters_birdnet = {
   "model" : "birdnet",
   "output_path" : "outputs",
   "filetype" : "wav", 
   "num_predictions" : 1,
   "write_to_csv" : True
}
automated_df_birdnet,local_scores = generate_automated_labels(path_to_audio_files,isolation_parameters_birdnet);
automated_df_birdnet["MANUAL ID"] = "bird"
automated_df_birdnet_chunked = annotation_chunker(automated_df_birdnet, chunk_length)
```

```python
isolation_parameters_tweety = {
   "model" : "tweetynet",
    "tweety_output" : True,
    "chunk_size" : 3.0
}
automated_df_tweety = generate_automated_labels(path_to_audio_files,isolation_parameters_tweety);
automated_df_tweety["MANUAL ID"] = "bird"
automated_df_tweety_chunked = annotation_chunker(automated_df_tweety, chunk_length)
```

Compared Chunk Results to Non Chunked Manual Data

```python
statistics_df_tweety = automated_labeling_statistics(automated_df_tweety,manual_df,stats_type = "general");
statistics_df_micro = automated_labeling_statistics(automated_df_birdnet,manual_df,stats_type = "general");
statistics_df_birdnet = automated_labeling_statistics(automated_df_micro,manual_df,stats_type = "general");
statistics_df_micro_chunked = automated_labeling_statistics(automated_df_micro_chunked,manual_df,stats_type = "general");
statistics_df_birdnet_chunked = automated_labeling_statistics(automated_df_birdnet_chunked,manual_df,stats_type = "general");
statistics_df_tweety_chunked = automated_labeling_statistics(automated_df_tweety_chunked,manual_df,stats_type = "general");
```

```python
display(global_dataset_statistics(statistics_df_micro).style.set_caption("Microfaune Annotations"))
display(global_dataset_statistics(statistics_df_birdnet).style.set_caption("BirdNET Annotations"))
display(global_dataset_statistics(statistics_df_tweety).style.set_caption("TweetyNET Annotations"))
display(global_dataset_statistics(statistics_df_micro_chunked).style.set_caption("Microfaune Annotations chunked"))
display(global_dataset_statistics(statistics_df_birdnet_chunked).style.set_caption("BirdNET Annotations chunked"))
display(global_dataset_statistics(statistics_df_tweety_chunked).style.set_caption("TweetyNET Annotations chunked"))
```

```python
generate_ROC_curves(automated_df_tweety, manual_df, chunk_length = 3, label="tweetynet")
generate_ROC_curves(automated_df_micro_chunked, manual_df, chunk_length = 3, label="mircofaune chunked")
generate_ROC_curves(automated_df_birdnet_chunked, manual_df, chunk_length = 3, label="birdnet chunked")
generate_ROC_curves(automated_df_tweety_chunked, manual_df, chunk_length = 3, label="tweety chunked")
```
